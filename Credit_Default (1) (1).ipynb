```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train_credit.csv")
```

```python
print(data.columns)
```

```python
print(data.describe())
```

```python
data.hist(figsize=(20,20))
plt.show()
```

```python
default = data[data['default_payment_next_month']==1]
not_default = data[data['default_payment_next_month']==0]

outlier_fraction = len(default)/float(len(not_default))
print(outlier_fraction)

print("Default Cases:{}".format(len(default)))
print("Non Default Cases:{}".format(len(not_default)))
```

```python
corrmat = data.corr()
fig=plt.figure(figsize=(12,9))

sns.heatmap(corrmat , vmax=1 , square=True)
plt.show()
```

```python
data.drop(['ID'],axis=1)
```

```python
columns = data.columns.tolist()

columns = [c for c in columns if c not in ["default_payment_next_month","ID"]]
           
target = "default_payment_next_month"           
X_train=data[columns]
Y_train=data[target]
print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,random_state=0)
classifier.fit(X_train,Y_train)
```

```python
data_test = pd.read_csv('test.csv')
```

```python
data_test.head
```

```python
y_pred = classifier.predict(data_test)
```

```python
y_pred
```

```python
pred = pd.DataFrame(y_pred)
sub_df = pd.read_csv('sample_submission.csv')
datasets = pd.concat([sub_df['ID'],pred],axis=1)
datasets.columns = ['ID','default_payment_next_month']
datasets.to_csv('sample_submission_send.csv',index=False)
```
